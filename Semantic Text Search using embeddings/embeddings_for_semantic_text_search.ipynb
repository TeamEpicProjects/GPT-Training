```python
import pandas as pd
import tiktoken
import openai
from openai.embeddings_utils import get_embedding
```

```python
openai.api_key = "Enter your OpenAI API Key here."
```

```python
# embedding model parameters
embedding_model = "text-embedding-ada-002"
embedding_encoding = "cl100k_base"  # this the encoding for text-embedding-ada-002
max_tokens = 8000  # the maximum for text-embedding-ada-002 is 8191
```

```python
file_path = 'dataset/data.csv'

df = pd.read_csv(file_path)
```

```python
df
```

```python
df_sample = df.groupby('title')
```

```python
df_sample.first().reset_index()
```

```python
df_sample=df.groupby('title').first().reset_index()
```

```python
df_sample
```

```python
df_sample.isna().sum()
```

```python
df_sample['combined'] = ("Summary: " + df.summary.str.strip() + "; Content: " + df.reviewText.str.strip())
```

```python
df_sample=df_sample.dropna()
```

```python
encoding = tiktoken.get_encoding(embedding_encoding)
```

```python
df_sample['n_tokens'] = df_sample.combined.apply(lambda x: len(encoding.encode(x)))
```

```python
df_sample.isna().sum()
```

```python
df_sample = df_sample[df_sample.n_tokens <= max_tokens].tail(1000)
```

```python
len(df_sample)
```

```python
df_sample["embedding"] = df_sample.combined.apply(lambda x: get_embedding(x, engine=embedding_model))
```

```python
df_sample.to_csv('dataset/sampledata_with_embeddings.csv', index=False)
```
